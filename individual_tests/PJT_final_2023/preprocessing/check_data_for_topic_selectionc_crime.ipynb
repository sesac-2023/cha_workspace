```python
import pandas as pd
```

```python
# 컬럼정의서에서 컬럼명 추출하기
with open('KPU_26_컬럼정의서_20230925.csv', 'r', encoding='cp949') as f:
    column = [d.rstrip() for d in f.readlines()]
column_name = []
for i in column:
    column_name.append(i.split(',')[2].strip())
column_name = column_name[1:]
```

```python
# 컬럼이 없는 데이터에 컬럼명 추가하기
df = pd.read_csv('KPU_26_20211231_C_001.csv')
df.columns = column_name
df
```

https://bigdata-policing.kr/product/view?product_id=PRDT_505

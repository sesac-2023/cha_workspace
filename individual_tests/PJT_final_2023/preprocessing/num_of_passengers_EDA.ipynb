```python
import pandas as pd

data = pd.read_csv('./data/버스_지하철_승하차인원_8월평균.csv', index_col=0)

# int로 타입 변환
data = data.astype('int')

# 총 승하차 인원 내림차순으로 재정렬
df = data.sort_values(by=['승하차인원_버스_지하철'], ascending=False).head(10)
df
```

버스+지하철 승하차인원 상위 10개 행정동 추출

```python
import seaborn as sns
import matplotlib.pyplot as plt

f, ax = plt.subplots(figsize=(4, 4))

sns.set(style="whitegrid", font='NanumGothic', font_scale=0.8)
sns.set_color_codes("pastel")
sns.barplot(x='승하차인원_버스_지하철', y=df[:5].index, data=df[:5],
            label="버스+지하철", color="g")

sns.set_color_codes("muted")
sns.barplot(x='승하차인원_지하철', y=df[:5].index, data=df[:5],
            label="지하철", color="g")

ax.legend(ncol=2, loc="lower right", frameon=True)
ax.set(ylabel="",xlim=(0, 2000000))
sns.despine(left=True, bottom=True)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

f, ax = plt.subplots(figsize=(4, 4))

sns.set(style="whitegrid", font='NanumGothic', font_scale=0.8)
sns.set_color_codes("pastel")
sns.barplot(x='승하차인원_버스_지하철', y=df[5:].index, data=df[5:],
            label="버스+지하철", color="g")

sns.set_color_codes("muted")
sns.barplot(x='승하차인원_지하철', y=df[5:].index, data=df[5:],
            label="지하철", color="g")

ax.legend(ncol=2, loc="lower right", frameon=True)
ax.set(ylabel="",xlim=(0, 2000000))
sns.despine(left=True, bottom=True)
```

```python
f, ax = plt.subplots(1, 2, figsize=(8, 4))

for idx, df in enumerate([df[:5], df[5:]]):
    ax_idx = sns.barplot(x='승하차인원_버스_지하철', y=df.index, data=df, palette="crest", ax=ax[idx])
    ax_idx.set(xlim=(0,2000000))
plt.show()
```

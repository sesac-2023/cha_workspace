```python
import pandas as pd
```

```python
df1 = pd.read_csv("서울시 역사마스터 정보.csv", usecols=['역사명', '위도', '경도'])
df2 = pd.read_csv('./서울교통공사_지하상가공실.csv', usecols=['역명', '상가번호', '면적(제곱미터)', '사업진행단계'])
```

```python
df1.head(2)
```

```python
df2.head(2)
```

```python
df1['_역명'] = df1['역사명'].map(lambda x : x.split('역')[0].split('(')[0]).drop_duplicates()
df2['_역명'] = df2['역명'].map(lambda x : x.split('역')[0].split('(')[0])
```

```python
df3 = pd.merge(df2, df1, on='_역명', how='inner')
```

```python
df3[:10]
```

```python
df3.to_csv('test.csv', columns=['역사명', '상가번호', '면적(제곱미터)', '사업진행단계', '위도', '경도'], index=False)
```

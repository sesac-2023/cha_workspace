기준금리 월 단위 라벨링

```python
import pandas as pd
import datetime
```

```python
data = pd.read_csv('./한국은행_기준금리_최대값 [2014.01~2023.08] (%).csv')
```

```python
df = data.copy()
```

```python
df
```

```python
month_to_int = {
    'January': 1,
    'February': 2,
    'March': 3,
    'April': 4,
    'May': 5,
    'June': 6,
    'July': 7,
    'August': 8,
    'September': 9,
    'October': 10,
    'November': 11,
    'December': 12
}

df['month'] = df['월'].map(month_to_int)
df.insert(0, 'date',"")
df['date'] = df['연도'].astype(str) + df['month'].astype(str).str.zfill(2) + '01'
```

```python
df
```

```python
df.drop(['연도', '분기', '월', 'month'], axis=1, inplace=True)
```

```python
df.rename(columns={'최대 값%': 'base_rate'}, inplace=True)
```

```python
df
```

```python
df['base_rate'].diff()
```

```python
df['up_down'] = df['base_rate'].diff()
```

```python
df
```

```python
df['up_down'] = df['up_down'].apply(lambda x: '상승' if x > 0 else ('하락' if x < 0 else '유지'))
```

```python
df
```

```python
df.to_csv('./labeled_base_rate.csv', index=None)
```

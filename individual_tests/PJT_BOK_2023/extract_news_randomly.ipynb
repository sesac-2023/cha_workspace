### 크롤링한 뉴스 기사 랜덤추출해서 살펴 보기

```python
from ekonlpy import Mecab
import pandas as pd
```

```python
df = pd.read_csv('./bok_hk_2022.txt')
df = df.dropna(axis=0).sample(n=100)
sorted_df = df.sort_values('date', ascending=False)
```

```python
sorted_df
```

```python
sorted_df.to_csv('read_news.txt', index=False)
```

```python
content_list = sorted_df['content'].tolist()
```

```python
content_list[0]
```

```python
mecab = Mecab()
for content in content_list:
    pos_tag = mecab.pos(content)
```

```python
pos_tag
```

```python
[word for word, pos in pos_tag if pos in ['NNG', 'NNP', 'VV', 'VA','VAX','MAG']]
```

```python
for i, j in pos_tag:
    print(i)
```

### matplotlib practice for DL
> activation function

- step function

```python
def step(x):
    return np.array(x > 0, dtype=int)

x = np.arange(-5.0, 5.0, 0.1)
y = step(x)
plt.title('Step Function')
plt.plot(x,y)
plt.show()
```

- sigmoid function

```python
def sigmoid(x):
    return 1/(1+np.exp(-x))

x = np.arange(-5.0, 5.0, 0.1)
y = sigmoid(x)

plt.plot(x, y)
plt.plot([0,0], [1.0, 0.0], ':') #   
plt.title('Sigmoid Function')
plt.show()
```

- Hyperbolic tangent function

```python
x = np.arange(-5.0, 5.0, 0.1)
y = np.tanh(x)
```

```python
plt.plot(x, y)
# plt.plot([0,0], [1.0,-1.0], ':')
plt.axhline(y=0, color='orange', linestyle='--') # horizontal
plt.axvline(x=0, color='orange', linestyle=':') # vertical
plt.title('Tanh Function')
plt.show()
```

- ReLU function

```python
def relu(x):
    return np.maximum(0, x)

x = np.arange(-5.0, 5.0, 0.1)
y = relu(x)

plt.plot(x, y)
plt.title('Relu function')
plt.axvline(x=0, color='orange', linestyle=':')
plt.show()
```

- Leaky ReLU function

```python
a = 0.1

def leaky_relu(x):
    return np.maximum(a*x, x)

x = np.arange(-5.0, 5.0, 0.1)
y = leaky_relu(x)

plt.plot(x, y)
plt.title('Leaky ReLU Function')
plt.plot([0,0], [5.0, 0.0], ':')
plt.show()
```

- softmax function

```python
x = np.arange(-5.0, 5.0, 0.1)
y = np.exp(x) / np.sum(np.exp(x))

plt.plot(x, y)
plt.title('Softmax Function')
plt.show()
```

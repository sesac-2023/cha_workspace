```python
# x ,y numpy array
W = np.random.rand(1, 1) # W
b = np.random.rand(1) # b

# Wx + b
print(W, b)
x_data = x.reshape(100, 1)
t_data = y.reshape(100, 1)

# np.dot -> 곱셈 기능
h = np.dot(x_data, W) + b

plt.scatter(x_data, t_data, label = "Data Scatter")
plt.plot(x_data, h, color="Red")
plt.legend()
plt.show()
```

```python
# h = Wx + b
# x, y
# hypothesis

# 오차의 합을 구한다. 기본 원리
# MSE - Mean Squared Error 오차제곱평균 -> 그대로 파이썬 수식으로
# np.dot -> numpy  행렬의 곱셈
# 원래 값들 ( y - (Wx + b) ) ** 2 -> 오차값을 제곱시킴 -> x의 개수로 나누면 오차제곱들의 평균

print( np.sum( ( y - (np.dot(x_data, W) + b) ) ** 2 ) / len(x) )
```

```python
def loss_function(x, t): # Loss Function 손실함수 정의
    y = np.dot(x, W) + b
    return ( np.sum( (t-y)**2 ) / (len(x)) )

learning_rate = 1e-5
f = lambda x : loss_function(x_data, t_data)
print(f"Initial Loss Value : {loss_function(x_data, t_data)}\nInitial W : {W}\nInitial b : {b}")
```

```python
def derivative(f, var):
    delta = 1e-5
    grad = np.zeros_like(var)
    it = np.nditer(var, flags=["multi_index"], op_flags=["readwrite"])
    # print(x)
    # print(var)
    while not it.finished:

        idx = it.multi_index
        tmp_val = var[idx]
        var[idx] = float(tmp_val) + delta
        fx1 = f(x)

        var[idx] = float(tmp_val) - delta
        fx2 = f(x)

        grad = ((fx1 - fx2) / (2*delta))
        var[idx] = tmp_val
        it.iternext()

    return grad
    # diff_val = (f_plus - f_minus) / (2*delta)
    # return diff_val

x_axis = []
y_axis = []
for step in range(1000):
    # print(W)
    # print(derivative(f, W))
    W -= learning_rate * derivative(f, W)
    # print(W)
    b -= learning_rate * derivative(f, b)
    if ( step % 10 == 0 ):
        print(W, b, loss_function(x_data, t_data))
        x_axis.append(W[0][0])
        y_axis.append(loss_function(x_data, t_data))
plt.plot(x_axis, y_axis, label='Curve')
plt.scatter(x_axis, y_axis, color='red', label='Data Points')  # 점들을 빨간색으로 표시
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Curve with Data Points')
plt.legend()
# 그래프 보여주기
plt.show()
```

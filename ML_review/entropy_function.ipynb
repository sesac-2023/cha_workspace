## Entropy Funtion
- 케이스가 많아서 예측이 어려울 수록 엔트로피 지수가 높다.

### sigmoid 함수 적용

```python
import numpy as np
import matplotlib.pyplot as plt

x_1 = np.linspace(-1.0, 1.0, 10).reshape(10, 1)
x_2 = np.linspace(-1.0, 1.0, 10).reshape(10, 1)

w_1 = [[2.0]]
w_2 = [[3.0]]

bias = 0.1

output_data = np.zeros((10, 10))

for i in range(10):
    for j in range(10):
        h = x_1[i]*w_1 + x_2[j]*w_2 + bias
        z = 1 / ( 1 + np.exp(-h) ) # sigmoid 함수
        output_data[i][j] = z

plt.imshow(output_data, "gray", vmin=0.0, vmax=1.0)
plt.colorbar()
plt.show()
```

- 가중치 변화 살펴 보기

```python
w_1 = [[10]]
w_2 = [[20]]

bias = 100

output_data = np.zeros((10, 10))

for i in range(10):
    for j in range(10):
        h = x_1[i]*w_1 + x_2[j]*w_2 + bias
        z = 1 / ( 1 + np.exp(-h) )
        output_data[i][j] = z

plt.imshow(output_data, "gray", vmin=0.0, vmax=1.0)
plt.colorbar()
plt.show()
```

### tanh 함수 적용

```python
w_1 = [[10]]
w_2 = [[20]]

bias = 100

output_data = np.zeros((10, 10))

for i in range(10):
    for j in range(10):
        h = x_1[i]*w_1 + x_2[j]*w_2 + bias
        z = ( np.exp(h) - np.exp(-h) ) / ( np.exp(h) + np.exp(-h) ) # tanh 함수
        output_data[i][j] = z

plt.imshow(output_data, "gray", vmin=0.0, vmax=1.0)
plt.colorbar()
plt.show()
```

### Relu 함수 적용

```python
w_1 = [[10]]
w_2 = [[0.5]]

bias = 100

output_data = np.zeros((10, 10))

for i in range(10):
    for j in range(10):
        h = x_1[i]*w_1 + x_2[j]*w_2 + bias
        z = np.where(h<=0, 0, h)  # Relu 함수
        output_data[i][j] = z

plt.imshow(output_data, "gray", vmin=0.0, vmax=1.0)
plt.colorbar()
plt.show()
```

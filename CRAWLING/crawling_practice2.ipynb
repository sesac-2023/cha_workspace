네이버 주가 크롤링
- https://finance.naver.com/sise/sise_quant.nhn
1. 종목명과 현재가를 크롤링해주세요.
2. 전일대비 상승한 종목만 종목명, 현재가, 전일비를 크롤링해주세요

```python
import requests
from bs4 import BeautifulSoup
headers = {'User-Agent' :"Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/114.0.0.0 Safari/537.36"}

url = "https://finance.naver.com/sise/sise_quant.nhn"
response = requests.get(url, headers=headers)
response
```

```python
soup = BeautifulSoup(response.text, 'html.parser')
trTags = soup.select('.type_2 tr')  #type_2 class 의 후손 셀렉터 tr
```

방법 1. slicing (start:end:step)

```python
# 1. slicing (start:end:step)
for i in range(20):
    targets = trTags[2:][8*i:8*i+5]
    print([t.select_one('.tltle').text.strip() for t in targets])
```

방법 2. try - except
방법 3. if - else

```python
result = []
upStocks = []
for trTag in trTags[2:]:
    try:
        print(trTag.select_one('.tltle').text.strip())
        temp = {
            '종목명' : trTag.select_one('.tltle').text.strip(),
            '현재가' : trTag.select_one('.number').text.strip()
        }
        # 상승, 하락 이미지가 있는지 확인 -> alt 속성 체크
        if trTag.select_one('img'):
            if 'ico_up' in trTag.select_one('img').attrs['src']:
                upStocks.append({
                    '종목명' : trTag.select_one('.tltle').text.strip(),
                    '현재가' : trTag.select_one('.number').text.strip(),
                    '전일비' : trTag.select('.number')[1].text.strip()

                }) 
        result.append(temp)
    except:
        pass
   
```

무신사 쇼핑몰 크롤링
- 무신사 쇼핑몰 상의(TOP) 카테고리 첫페이지 제품들의 브랜드명, 제품명, 원래가격, 할인가격을 크롤링해주세요.
- https://store.musinsa.com/app/items/lists/001
- 성공하신분들은 총 10페이지까지 크롤링해주세요.

```python
url = "https://www.musinsa.com/categories/item/001"
response = requests.get(url, headers=headers)
response
```

```python
soup = BeautifulSoup(response.text, 'html.parser')
items = soup.select('#searchList .article_info')
```

```python
musinsa = []
for item in items:
    price = item.select_one('.price').text.strip().split('\n')

    musinsa.append({
        '브랜드명' : item.select_one('.item_title').text.strip(),
        '상품명' : item.select_one('.list_info').text.strip(),
        '정상 가격' : price[0].strip(),
        '할인 가격' : price[-1].strip()     # 할인 가격이 없는 경우 price[1]로 설정시 오류 -> 1) if문으로 해결 or 2) price[-1]로 해결 가능
    })

musinsa
```

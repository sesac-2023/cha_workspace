```python
import requests
from bs4 import BeautifulSoup
headers = {'User-Agent' :"Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/114.0.0.0 Safari/537.36"}

url = "https://finance.naver.com/sise/sise_quant.nhn"
response = requests.get(url, headers=headers)
response
```

```python
soup = BeautifulSoup(response.text, 'html.parser')
trTags = soup.select('.type_2 tr')  #type_2 class 의 후손 셀렉터 tr
```

방법 1. slicing (start:end:step)

```python
# 1. slicing (start:end:step)
for i in range(20):
    targets = trTags[2:][8*i:8*i+5]
    print([t.select_one('.tltle').text.strip() for t in targets])
```

방법 2. try - except
방법 3. if - else

```python
result = []
upStocks = []
for trTag in trTags[2:]:
    try:
        print(trTag.select_one('.tltle').text.strip())
        temp = {
            '종목명' : trTag.select_one('.tltle').text.strip(),
            '현재가' : trTag.select_one('.number').text.strip()
        }
        # 상승, 하락 이미지가 있는지 확인 -> alt 속성 체크
        if trTag.select_one('img'):
            if 'ico_up' in trTag.select_one('img').attrs['src']:
                upStocks.append({
                    '종목명' : trTag.select_one('.tltle').text.strip(),
                    '현재가' : trTag.select_one('.number').text.strip(),
                    '전일비' : trTag.select('.number')[1].text.strip()

                }) 
        result.append(temp)
    except:
        pass
   
```

> 방법 2 - 페이지 10까지 크롤링하는 함수 생성

1. url 가져오기
2. '?'기준으로 베이스 url과 쿼리파라미터 구분하기 ('?' 삭제)
3. 쿼리파라미터 중에 값이 없는 요소 삭제하기
4. 값이 있는 쿼리파라미터를 params 딕셔너리에 담기
5. params의 key에 접근해서 page 값을 바꾸는 반복문 실행

```python
url = 'https://www.musinsa.com/categories/item/001'

# d_cat_cd=001
# list_kind=small
# sort=pop_category
# page=1
# display_cnt=90
```

```python
response = requests.get(url, headers=headers)
response
```

```python
def musinsa_crawl(response):
    soup = BeautifulSoup(response.text, 'html.parser')
    items = soup.select('#searchList .article_info')

    musinsa = []
    for item in items:
        price = item.select_one('.price').text.strip().split('\n')

        musinsa.append({
            '브랜드명' : item.select_one('.item_title').text.strip(),
            '상품명' : item.select_one('.list_info').text.strip(),
            '정상 가격' : price[0].strip(),
            '할인 가격' : price[-1].strip()     # 할인 가격이 없는 경우 price[1]로 설정시 오류 -> 1) if문으로 해결 or 2) price[-1]로 해결 가능
        })

    return musinsa  # -> list
```

```python
params =  {
    'd_cat_cd' : '001',
    'list_kind' : 'small',
    'sort' : 'pop_category',
    'page' : 1,
    'display_cnt' : 90
}
```

```python
result = []
for page in range(1, 11):
    params['page'] = page  # params 딕셔너리의 key에 접근해서 value 변경하기

    requests.get(url, params=params)    # headers=headers 대신 params=params
    musinsa = musinsa_crawl(response)

    result.extend(musinsa)
```

```python
# soup = BeautifulSoup(response.text, 'html.parser')

def news_crawl(soup):
    newsItems = soup.select('ul[class^="type06_headline"] li')
    result = []
    for news in newsItems:
        title = news.select('dt')[-1].text.strip()
        content = news.select_one('.lede').text.strip()

        result.append({
            '제목' : title,
            '요약' : content
        })
    return result
```

```python
page = 1

beforeNews = []

result = []
while True:
    print(page)
    response = requests.get(url.format('20230713', page), headers=headers)
    soup = BeautifulSoup(response.text, 'html.parser')
    news = news_crawl(soup)

    page += 1

    if beforeNews == news:    # 이전 뉴스 페이지가 현재 뉴스 페이지와 동일하면 중지
        break
    else:
        beforeNews = news     # 이전 뉴스를 최신 뉴스로 업데이트

    result.extend(news)

result
```

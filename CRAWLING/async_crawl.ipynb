네이버 웹툰 크롤링
- 제목과 썸네일 이미지 url 가져오기
- https://comic.naver.com/webtoon/weekday?order=User

```python
# url = 'https://comic.naver.com/api/webtoon/titlelist/weekday?order=view'
url = 'https://comic.naver.com/api/webtoon/titlelist/weekday?order=user'  # 개발자 도구 - 네트워크탭 
response = requests.get(url)
response
```

```python
data = json.loads(response.text)
```

```python
# 딕셔너리 자료형인 data의 output 일부부만 출력하기

for key, item in dict(data).items():
    print(key,item)
    break
```

```python
# data output
#  {'titleListMap': {'WEDNESDAY': [{'titleId': 769209,
#     'titleName': '화산귀환',
#     'author': 'LICO / 비가',
#     'thumbnailUrl': 'https://image-comic.pstatic.net/webtoon/769209/thumbnail/thumbnail_IMAG21_3511dcdd-6e33-4171-8839-598d6d266215.jpg',
#     'up': False,
#     'rest': False,
#     'bm': False,
#     'adult': False,
#     'starScore': 9.92877,
#     'viewCount': 0,
#     'openToday': False,
#     'potenUp': False,
#     'bestChallengeLevelUp': False,
#     'finish': False,
#     'new': False},
```

```python
webtoons = []
for key in data['titleListMap']:  # 요일
    for webtoon in data['titleListMap'][key]:   # 해당 요일의 웹툰 정보들
        webtoon['titleName']   # 제목
        webtoon['thumbnailUrl']  # 썸네일 url

        webtoons.append({
            '제목' : webtoon['titleName'],
            '썸네일' :  webtoon['thumbnailUrl']
        })

        if len(webtoons) >= 5:
                break

    if len(webtoons) >= 5:
        break

webtoons
```

디자인정글 크롤링
- 제목, 카테고리
- 스크롤을 통해 나오는 추가 데이터도 크롤링
- https://www.jungle.co.kr/

```python
url = 'https://www.jungle.co.kr/recent.json'

# 쿼리파라미터
params = {
    'magazineOffset': 0,
    'contestOffset': 0,
    'exhibitOffset': 0,
    'galleryOffset': 0
}

response = requests.get(url, params=params)
data = json.loads(response.text)
```

```python
data
```

```python
result = []
while True:
    response = requests.get(url, params=params)
    data = json.loads(response.text)
    # for item in data['moreList']:
    #     params[item['targetName'] + 'Offset'] += 1  # 데이터의 params값이 주어지지 않을 때
    for key in params:
        params[key] = data[key]
    print(params)    

    # to do! break 문 추가하기
  
    for item in data['moreList']:
        result.append({
            '카테고리' : item['targetCode'],
            '제목' : item['title']
        })

    if len(result) >= 10:
        break
```

```python
def crawl_rocketpunch(page):
    response = requests.get(url.format(page))
    data = json.loads(response.text)
    html = data['data']['template']
    soup = BeautifulSoup(html, 'html.parser')
    company_list = soup.select('.company.item')

    rocket_result = []
    for company in company_list:
        company.select_one('.company-name').text.strip()
        company.select_one('.description').text.strip()
        [job.text.strip() for job in company.select('.job-name')]

        rocket_result.append({
            '회사명': company.select_one('.company-name').text.strip(),
            '설명': company.select_one('.description').text.strip(),

            # '공고들': [(job.text.strip(), job.attrs['href']) for job in company.select('.job-name > a:first-child')],
            '공고들': [(job.text.strip(), job.select_one('a').attrs['href']) for job in company.select('.job-name')],
        })

    return rocket_result
```

```python
rocket_result = []
for page in range(1, 3):
    rocket_result.extend(crawl_rocketpunch(page))

    if len(rocket_result) >= 1:
        break

rocket_result[:5]
```

## 실습 3
- 2017.12.1.xlsx ~ 2017.12.9.xlsx 엑셀 파일의 내용을 하나의 엑셀 파일에 취합

### ANSWER

```python
import os

FILE_PATH = './실습3/'
# 1. os.listdir()로 엑셀 파일 가져오기
excel_files = os.listdir(FILE_PATH)
print(excel_files)

log_excel = openpyxl.Workbook()
log_sheet = log_excel.active

# 2. 엑셀 파일을 하나씩 열어서 각 cell의 값 확인
for excel_file in excel_files:
    print(excel_file)
    wb = openpyxl.load_workbook(FILE_PATH + excel_file)
    ws = wb.active
    cells = ws['A1:C24']

    for row in cells:
        print(row)
        values = [cell.value for cell in row]

        # 1. append
        log_sheet.append(values)

        # 2. row index
        row_index = 1
        for log_row in log_sheet[f'A{row_index}:C{row_index}']:
            for i, cell in enumerate(log_row):
                print(cell)
                cell.value = row[i].value
        row_index += 1


# 3. 데이터 저장 or 바로 새 workbook에 쓰기
log_excel.save(FILE_PATH + 'log.xlsx')
```
